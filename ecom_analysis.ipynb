```python
import pandas as pd
```

```python
import psycopg2
from sqlalchemy import create_engine
```

```python
order_df.head()
```

```python
aisles_df = pd.read_csv("F:\\csv file\\Data\\aisles.csv")
departments_df = pd.read_csv("F:\\csv file\\Data\\departments.csv")
order_products_df = pd.read_csv("F:\\csv file\\Data\\order_products.csv").sample(10000)
products_df = pd.read_csv("F:\\csv file\\Data\\products.csv")
```

```python
try:
    conn = psycopg2.connect(dbname='ecom_analysis',user='postgres',password='root',port='5432')
except:
    print("connection was unsuccessful")
```

```python
engine = create_engine('postgresql+psycopg2://postgres:root@localhost/ecom_analysis')
```

```python
curr = conn.cursor()
```

```python
curr.execute("""
CREATE TABLE aisles(
    aisle_id INTEGER PRIMARY KEY,
    aisle VARCHAR
)
""")
```

```python
curr.execute("""
CREATE TABLE departments(
    department_id INTEGER PRIMARY KEY,
    department VARCHAR(255)
)
""")
```

```python
curr.execute("""
CREATE TABLE products (
    product_id INTEGER PRIMARY KEY,
    product_name VARCHAR(255),
    aisle_id INTEGER,
    department_id INTEGER,
    FOREIGN KEY (aisle_id) REFERENCES aisles (aisle_id),
    FOREIGN KEY (department_id) REFERENCES departments (department_id)
)
""")
```

```python
curr.execute("""
CREATE TABLE orders (
    order_id INTEGER PRIMARY KEY,
    user_id INTEGER,
    order_number INTEGER,
    order_dow INTEGER,
    order_hour_of_day INTEGER,
    days_since_prior_order INTEGER )
""")
```

```python
curr.execute("""
CREATE TABLE order_products (
    order_id INTEGER,
    product_id INTEGER,
    add_to_cart_order INTEGER,
    reordered INTEGER,
    PRIMARY KEY (order_id, product_id),
    FOREIGN KEY (order_id) REFERENCES orders (order_id),
    FOREIGN KEY (product_id) REFERENCES products (product_id)
    )
""")
```

```python
conn.commit()
```

```python

products_df.to_sql('products', con=engine, if_exists='append', index=False)
```
